Using List to find most similar post in CH3 Thailand Posts

Then, find the top 10 most similar posts. The similarity of two posts can be determined by <BR>

$similarity(wordset_1,wordset_2)=\frac{|intersection(wordset_1,wordset_2)|}{|union(wordset_1,wordset_2)|}$

where $|s|$ means the number of members in $s$. For example, if $s_1$={'วันนี้','ฉัน','ไป','กิน','ข้าว'} and $s_2$={'กิน','ข้าว','ไม่','อร่อย'}, the similarity of $s_1$ and $s_2$ will be $\frac{2}{7}$.

```python
from pythainlp import word_tokenize 

def get_posts():
  # readlines
  file=open('CH3Thailand Posts.csv')
  lines=file.readlines()
  file.close 

  # remove hashtags 
  lines=[' '.join([sentence for sentence in line.strip().split() if sentence[0] != '#']) for line in lines]

  # Thai words tokenization
  posts=[tuple(word_tokenize(line,engine='newmm',keep_whitespace=False)) for line in lines]
  return posts
  
posts=get_posts()
```

```python
def find_most_similar_pairs(posts):
  sims=[]
  for i in range(0,len(posts)):
    for k in range(i+1,len(posts)):
      pi=set(posts[i])
      pk=set(posts[k])
      if pi != pk:
        s = len(pi.intersection(pk))/len(pi.union(pk))
        sims.append((s,posts[i],posts[k]))
  return sorted(sims, reverse=True)

similar_pairs=find_most_similar_pairs(posts)
```

```python
def show_posts(p):
  ret=' '.join(p)
  if len(ret) > 30:
    return ret[:30]+'...'
  return ret

def show_result(top_similar_pairs):
  for t in top_similar_pairs:
    print(t[0],',',show_posts(t[1]),show_posts(t[2]))
  
show_result(similar_pairs[:100])
```
