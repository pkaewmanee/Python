```python
# Write your code here
lines = open('26-Aug-2021.csv').readlines()
data = [[e for e in line.split(',')] for line in lines[1:]]
n = sorted([int(n) for [dt,n] in data], reverse=True)
dt = sorted([dt for [dt,n] in data], reverse=True)
for i in range(len(data)):
  print(data[i])
  print(n[i])
  print(dt[i])
```

Make a graph out of a list

```python
import matplotlib.pyplot as plt 
import matplotlib.dates as mdates
from matplotlib import dates
from datetime import datetime 

# Write your code here
X=[dt for [dt,n] in data]
Y=[int(n) for [dt,n] in data]


X=([datetime.strptime(day, '%Y-%m-%d') for day in X])
plt.figure(figsize=(16,8))
plt.plot(X,Y)
plt.xticks(rotation=90)
plt.show()
```

Graph with both actual and 3-day moving averaged number of confirmed cases

```python
import matplotlib.pyplot as plt 
import matplotlib.dates as mdates
from matplotlib import dates
from datetime import datetime 

# Write your code here
X=[dt for [dt,n] in data]
Y=[int(n) for [dt,n] in data]
Z=[(Y[i-1]+Y[i]+Y[i+1])/3 if 0<i<(len(Y)-1) else Y[i] for i in range(len(Y))]

X=([datetime.strptime(day, '%Y-%m-%d') for day in X])
plt.figure(figsize=(16,8))
plt.plot(X,Y)
plt.plot(X,Z)
plt.xticks(rotation=90)
plt.show()
```
