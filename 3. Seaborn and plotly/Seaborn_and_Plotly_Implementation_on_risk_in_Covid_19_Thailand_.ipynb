# Implementation on risk in Covid-19 Thailand

1. Read the downloaded data file.
2. Convert the 'announce_date' to numeric
3. Create your new dataframe that can count all confirmed cases on each day
4. Plot your graph colored by the risk

```python
! wget https://data.go.th/dataset/8a956917-436d-4afd-a2d4-59e4dd8e906e/resource/026c8d31-0d51-49b7-9dc2-8373e7cef440/download/confirmed-cases-since-120864.csv
```

```python
# Write your code here 
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

df=pd.read_csv('confirmed-cases-since-120864.csv')

df['announce_date']=pd.to_datetime(df['announce_date'], dayfirst=True)
df=df.groupby(['announce_date','risk'])['No.'].count()
df=df.reset_index()
fig=px.line(df,x=df['announce_date'],y=df['No.'],color='risk')
fig.show()
```
